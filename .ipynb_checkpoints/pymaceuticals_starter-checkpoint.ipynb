## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from sklearn import datasets

# Study data files
mouse_metadata = "Mouse_metadata.csv"
study_results = "Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset
combined_mouse_data = pd.merge(mouse_metadata, study_results, on="Mouse ID")
# Set the 'Mouse ID' to be our index for easy referencing of rows
#combined_mouse_data = combined_mouse_data.set_index("Mouse ID")
combined_mouse_data.head()
```

## Summary statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
#and SEM of the tumor volume for each regimen
combined_mouse_data.describe()
```

```python
df_mouse = pd.DataFrame(combined_mouse_data)
df_mouse
```

```python
# table of mean, median, variance, standard deviation, and SEM
average = combined_mouse_data["Tumor Volume (mm3)"].mean()
#average
variance = combined_mouse_data["Tumor Volume (mm3)"].var()
#variance
median = df_mouse.loc[:,"Tumor Volume (mm3)"].median()
#median
std = df_mouse.loc[:,"Tumor Volume (mm3)"].std()
#std
sem = df_mouse.loc[:,"Tumor Volume (mm3)"].sem()
#sem
sum_stats = pd.DataFrame({"Mean": [average],
                                   "Median": [median],
                                   "Variance": [variance],
                                   "Standard Deviation": [std],
                                   "SEM": [sem]
                                   })
sum_stats
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

sumstats = pd.DataFrame(combined_mouse_data.groupby("Drug Regimen").count())
sumstats["Mean"] = pd.DataFrame(combined_mouse_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean())
sumstats["Median"] = pd.DataFrame(combined_mouse_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].median())
sumstats["Standard Deviation"] = pd.DataFrame(combined_mouse_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].std())
sumstats["Variance"] = pd.DataFrame(combined_mouse_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].var())
sumstats["SEM"] = pd.DataFrame(combined_mouse_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem())
sumstats = sumstats[["Mouse ID", "Mean", "Median", "Standard Deviation", "Variance", "SEM"]]
sumstats = sumstats.rename(columns = {"Mouse ID" : "Trials"})

sumstats.head()
```

## Quartiles, outliers and boxplots

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_grouped_df = df_mouse.groupby(['Drug Regimen','Timepoint'])['Tumor Volume (mm3)']
avg_tumor_df = tumor_grouped_df.mean()

# Convert to DataFrame
avg_tumor_df = avg_tumor_df.reset_index()

# Preview DataFrame
avg_tumor_df.head()
```

```python

# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_sem = tumor_grouped_df.sem()

# Convert to DataFrame
tumor_sem_df = pd.DataFrame(tumor_sem)

# Preview DataFrame
tumor_sem_df.reset_index(inplace=True)
tumor_sem_df.head()
```

```python

# Minor Data Munging to Re-Format the Data Frames
pivot_avg_tumor_df = avg_tumor_df.pivot(index='Timepoint',columns='Drug Regimen',values='Tumor Volume (mm3)')

# Preview that Reformatting worked
pivot_avg_tumor_df.head(20)
```

```python
max_timepoint=df_mouse.groupby(['Mouse ID']).max()
max_timepoint=max_timepoint.reset_index()
max_timepoint
```

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. 
max_timepoint=max_timepoint[['Mouse ID','Timepoint']].merge(df_mouse,on=['Mouse ID','Timepoint'],how='left')
Capomulin_df=max_timepoint.loc[max_timepoint['Drug Regimen']=='Capomulin']['Tumor Volume (mm3)']
Infubinol_df=max_timepoint.loc[max_timepoint['Drug Regimen']=='Infubinol']['Tumor Volume (mm3)']
Ramicane_df=max_timepoint.loc[max_timepoint['Drug Regimen']=='Ramicane']['Tumor Volume (mm3)']
Ceftamin_df=max_timepoint.loc[max_timepoint['Drug Regimen']=='Ceftamin']['Tumor Volume (mm3)']

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
mark_outlier=dict(markerfacecolor='red',markersize=10) 
# Generate a box & whisker plot of the final tumor volume of each mouse across four regimens of interest
# Highlight outliers by changing color and style
plt.boxplot([Capomulin_df,Ramicane_df, Infubinol_df, Ceftamin_df],labels=['Capomulin','Ramicane','Infubinol','Ceftamin'],
            flierprops=mark_outlier)
plt.title('Tumor Volume')
plt.xlabel('Drug')
plt.ylabel('Tumor Volume (mm3)')
plt.grid()
plt.show()
```

```python
# Generate the Plot 
drug_list = [('Capomulin','o','red'),('Infubinol','^','blue'),('Ketapril','s','green'),('Placebo','d','black')]
for drug,marker,colors in drug_list:
    ste = tumor_sem[drug]
    tumor_plt = plt.errorbar(pivot_avg_tumor_df.index,pivot_avg_tumor_df[drug],ste,
                                       fmt=marker,ls='--',color=colors,linewidth=0.5)

     
plt.legend(labels=['Capomulin','Ramicane','Infubinol','Ceftamin'], loc='best')
plt.title('Tumor Response')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')

plt.grid()

# Save the Figure
#plt.savefig(os.path.join('figures','tumor_response_to_treatment.png'))
```

## Line and scatter plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin

drug_list = [('Capomulin','o','red')]
for drug,marker,colors in drug_list:
    ste = tumor_sem[drug]
    tumor_plt = plt.errorbar(pivot_avg_tumor_df.index,pivot_avg_tumor_df[drug],ste,
                                       fmt=marker,ls='--',color=colors,linewidth=0.5)

     
plt.legend(['Capomulin'], loc='best')
plt.title('Tumor Response')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')

plt.grid()
```

```python
#mouse weight versus average tumor volume for the Capomulin
Cap_df = df_mouse.loc[(df_mouse["Drug Regimen"] == "Capomulin"),:]
Cap_df.head()
# tumor_weight_df = df_mouse.groupby(['Drug Regimen','Tumor Volume (mm3)'])['Weight (g)']
# avg_tumor_weight_df = tumor_weight_df.mean()

# Convert to DataFrame
#avg_tumor_weight_df = avg_tumor_weight_df.reset_index()

# Preview DataFrame
#avg_tumor_weight_df.head()
```

```python
# Pull values for x and y values
m_weight = Cap_df.groupby(Cap_df["Mouse ID"])["Weight (g)"].mean()
t_vol = Cap_df.groupby(Cap_df["Mouse ID"])["Tumor Volume (mm3)"].mean()

# Create Scatter Plot with values calculated above
plt.scatter(m_weight,t_vol)
plt.xlabel("Mouse Weight")
plt.ylabel("Tumor Volume")
plt.title('Mouse Weight vs Tumor Volume')
plt.grid()
plt.show()
```

```python
#Prep Data
m_weight = Cap_df.groupby(Cap_df["Mouse ID"])["Weight (g)"].mean()
t_vol = Cap_df.groupby(Cap_df["Mouse ID"])["Tumor Volume (mm3)"].mean()

# Linear regression model
slope, int, r, p, std_err = st.linregress(m_weight, t_vol)
fit = slope * m_weight + int

# Plot the linear model on top of scatter plot 
plt.scatter(m_weight,t_vol)
plt.xlabel("Mouse Weight")
plt.ylabel("Tumor Volume")
plt.title('Mouse Weight vs Tumor Volume')
plt.plot(m_weight,fit,"--")
plt.xticks(m_weight, rotation=90)
plt.show()

# Correlation coefficient
corr = round(st.pearsonr(m_weight,t_vol)[0],2)
print(f'The correlation between weight and tumor value is {corr}')
```
